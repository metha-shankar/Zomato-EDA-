```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how = 'left')
final_df.head()
```

```python
final_df.columns
```

```python
country_names = final_df['Country'].value_counts().index
country_names
```

```python
country_values = final_df['Country'].value_counts().values
country_values
```

```python
plt.pie(country_values[:3], labels = country_names[:3], autopct='%1.2f%%')
plt.show()
```

Observation:
  Zomato maximum transactions are from India followed by USA and UK

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

Observation:
 1. When rating is between 4.5 and 4.9 => Excellent
 2. When rating is between 4.0 and 4.4 => Very good
 3. When rating is between 3.5 and 3.9 => Good
 4. When rating is between 2.5 and 3.4 => Average
 5. When rating is between 1.8 and 2.4 => Poor
 6. When rating is 0 => Not rated

```python
ratings.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'Aggregate rating', y = 'Rating count', data = ratings, palette='viridis')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'Aggregate rating',
            y = 'Rating count',
            data = ratings,
            hue = 'Rating color',
            palette=['violet','red','orange','yellow','green','green']
            )
plt.show()
```

Observation:
1. No.of.customers who doesn't give any rating is very high
2. Max no.of.rating is between 2.5 and 3.4

```python
sns.countplot(x = 'Rating color', data = ratings, palette=['violet','red','orange','yellow','green','green'])
plt.show()
```

```python
# Find country names that has given 0 rating
zero_rated_countries = final_df[final_df['Aggregate rating'] == 0]['Country'].value_counts().reset_index()
zero_rated_countries
```

```python
plt.scatter(zero_rated_countries['Country'],
            zero_rated_countries['count'],
            marker ='o',
            c = ['purple', 'violet', 'brown', 'red'],
            s = [250, 150, 80, 50])
plt.show()
```

Observation: Zomato has the highest no.of.Indian customers who doesn't give any rating

```python
# Find out which currency used by which country
final_df[['Country','Currency']].drop_duplicates()
```

```python
# Find out the country that has highest no.of.transcations
final_df[['Country']].groupby('Country').size().reset_index().rename(columns={0:'Transaction count'})
```
